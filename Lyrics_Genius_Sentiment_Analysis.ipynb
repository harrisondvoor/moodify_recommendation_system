```python
# Load the dataset
df = pd.read_csv('/Users/hdvoor/Desktop/spotify_top_2023.csv', encoding='latin-1')
```

```python
# Preview the dataset
df.head()
```

```python
# Filter rows where artist_count is equal to 1
df = df[df['artist_count'] == 1]
```

```python
# Drop rows with non-ASCII characters instead of replacing the values
df = df[df['artist(s)_name'].apply(lambda x: all(ord(c) < 128 for c in x))]
df = df[df['track_name'].apply(lambda x: all(ord(c) < 128 for c in x))]
```

```python
# Function to fetch lyrics from Genius API
def get_lyrics(row):
    try:
        # Search for the lyrics using the artist and song title
        song = genius.search_song(row['track_name'], row['artist(s)_name'])
        # Return the lyrics if found
        return song.lyrics if song else None
    except Exception as e:
        print(f"Error fetching lyrics for {row['artist(s)_name']} - {row['track_name']}: {e}")
        return None
```

```python
# Apply the function to fetch lyrics and create a new 'lyrics' column
df['lyrics'] = df.apply(get_lyrics, axis=1)
```

```python
# Drop rows where lyrics couldn't be fetched
df = df.dropna(subset=['lyrics'])
```

```python
# Function to get mood categories based on lyrics sentiment
def get_mood(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0.4:
        return 'Ecstatic'
    elif 0.2 < analysis.sentiment.polarity <= 0.4:
        return 'Happy'
    elif 0 <= analysis.sentiment.polarity <= 0.2:
        return 'Okay'
    elif -0.2 <= analysis.sentiment.polarity < 0:
        return 'Sad'
    else:
        return 'Bummed'
```

```python
# Apply the mood analysis function to the 'lyrics' column and create a new 'mood' column
df['mood'] = df['lyrics'].apply(get_mood)
```

```python
# Drop column (because we don't need lyrics column)
df = df.drop(columns=['lyrics'])
```

```python
from sqlalchemy import create_engine, text
```

```python
# Pass the connection string to a variable, conn_url
conn_url = 'postgresql://postgres:123@localhost/SPOTIFY_MOOD_FINAL'
```

```python
# Create an engine that connects to PostgreSQL server
engine = create_engine(conn_url)
```

```python
# Establish a connection
connection = engine.connect()
```

```python
# Pass the SQL statements that create all tables
stmt = text("""
    CREATE TABLE spotify (
        song_id INT PRIMARY KEY,
        song_name VARCHAR(100),
        artist VARCHAR(100),
        artist_count INT,
        released_year INT,  
        released_month INT,
        released_day INT,
        in_spotify_playlists INT,
        in_spotify_charts INT,
        streams BIGINT,
        bpm INT,
        key CHAR(2),
        mode VARCHAR(10),
        danceability INT,
        valence INT,
        energy INT,
        acousticness INT,
        instrumentalness INT,
        liveness INT,
        speechiness INT,
        mood VARCHAR(50)
    );
""")
```

```python
# Execute the statement to create tables
connection.execute(stmt)
```
